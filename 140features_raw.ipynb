```python
import pandas as pd
from collections import OrderedDict
```

```python
df = pd.read_csv("/home/lzy/KDD-MIMIC/mimic_feature_set_c/resources/variable_map_feature_set_c/130features.csv")
itemids_df = df.query("table=='output'").item_id
```

```python
itemids = []
for i in itemids_df:
    itemids = itemids + i.split(",")
itemids = list(set(itemids))
itemids.remove("")
itemids = [x.strip() for x in itemids]
print(itemids)
```

```python
for i,row in df.iterrows():
    itemids_tmp = str(row["item_id"]).split(",")
    itemids_tmp = list(set(itemids_tmp))
    if "" in itemids_tmp:
        itemids_tmp.remove("")
    itemids_tmp = [x.strip() for x in itemids_tmp]
    for j in itemids_tmp:
        print(f'{j},{row["name"]}')
```
