```python
import pickle
def unit_coeffient():
    unit_co = {}
    paths = ["./resources/itemid_uom_map/inputevents_cv_map.csv",
            "./resources/itemid_uom_map/inputevents_mv_map.csv",]
    paths2 = ["./resources/itemid_uom_map/labevents_map.csv",]
    for p in paths:
        unit = pd.read_csv(p)
        for i, row in unit.iterrows():
            unit_co[ f'{row["ITEMID"]}-{row["AMOUNTUOM"]}' ] = row["COEFFICIENT"]
    for p in paths2:
        unit = pd.read_csv(p)
        for i, row in unit.iterrows():
            unit_co[ f'{row["ITEMID"]}-{row["VALUEUOM"]}' ] = row["COEFFICIENT"]
    # chartevents
    chart_uom = {227441: ['mg/dL', 'units'], 578:[ 'cmH20', '.'], 113:[ 'mmHg', '%'], 3451:[ 'kg', 'cm'],3723: ['kg', 'cm']}
    for i, v in chart_uom.items():
        unit_co[f'{i}-{v[0]}' ] = 0
        unit_co[f'{i}-{v[1]}' ] = 0
    with open("./resources/itemid_uom_map/unit_coeffient.pkl","wb") as f:
        pickle.dump(unit_co, f)
```

```python
unit = pd.read_csv("/home/lzy/KDD-MIMIC/mimic_benchmark/resources/itemid_uom_map/inputevents_cv_map.csv")
unit
```

```python
unit_co = {}
for i, row in unit.iterrows():
    unit_co[ f'{row["ITEMID"]}-{row["AMOUNTUOM"]}' ] = row["COEFFICIENT"]
```

```python
df_incv = df_incv.dropna(axis=0,subset = ["AMOUNT", "AMOUNTUOM"])
df_incv["AMOUNTUOM"] = df_incv["AMOUNTUOM"].str.lower()
for i,row in tqdm(df_incv.iterrows(),total = df_incv.shape[0]):
    strs = f'{row["ITEMID"]}-{row["AMOUNTUOM"]}'
    if not (strs in unit_co):
        continue
    num = unit_co[ strs ]
    if num !=0:
        df_incv.loc[i,"AMOUNT"] = row["AMOUNT"] / num
    else:
        df_incv.loc[i,"AMOUNT"] = 0
df_incv = df_incv.dropna(axis=0,subset = ["AMOUNT", "AMOUNTUOM"])
df_incv.shape
```

```python
df_incv = df_incv.dropna(axis=0,subset = ["AMOUNT", "AMOUNTUOM"])
# itemid with multiple uom
uom1 = df_incv.drop_duplicates(subset=["ITEMID","AMOUNTUOM"]).sort_values("ITEMID")
uom1 = uom1.groupby("ITEMID").count().reset_index()
itemid1 = uom1[uom1.AMOUNTUOM>1][["ITEMID"]]
itemid0 = uom1[uom1.AMOUNTUOM<=1][["ITEMID"]]
df_incv2 = df_incv.merge(itemid0,how="inner",left_on=["ITEMID"],right_on=["ITEMID"])
#df_incv2.shape
# itemid with 90%
uom2 = df_incv.merge(itemid1,how="inner",left_on=["ITEMID"],right_on=["ITEMID"])
uom2 = uom2.groupby(["ITEMID","AMOUNTUOM"]).count().reset_index()
uom2 = uom2.merge(uom2.groupby(["ITEMID"]).sum().reset_index(),how="inner",left_on="ITEMID",right_on="ITEMID")
uom2["Percent"] = uom2.AMOUNT_x / uom2.AMOUNT_y
uom2 = uom2[uom2.Percent>=0.9]
df_incv3 = df_incv.merge(uom2[["ITEMID","AMOUNTUOM"]],how="inner",left_on=["ITEMID","AMOUNTUOM"],right_on=["ITEMID","AMOUNTUOM"])
#df_incv3.shape
df_incv4 = pd.concat([df_incv2,df_incv3],axis=0).sort_values(["SUBJECT_ID","HADM_ID","ICUSTAY_ID"])
df_incv4.to_csv(os.path.join(mimic3_path2,"INPUTEVENTS_CV2.csv"))
print(df_incv4.shape)
print("- rows: ",(df_incv.shape[0]-df_incv4.shape[0]))
```

# OUTPUTEVENTS

```python
df_output = pd.read_csv(os.path.join(mimic3_path,"OUTPUTEVENTS.csv"),low_memory=False)
df_output
```

```python
AMOUNTS = df_output.VALUE.unique()
np.savetxt("./resources/others/amount_OUTPUTEVENTS.txt", AMOUNTS, fmt="%s")
```

```python
df_output = df_output.dropna(axis=0,subset = ["VALUE", "VALUEUOM"])
# itemid with multiple uom
uom1 = df_output.drop_duplicates(subset=["ITEMID","VALUEUOM"]).sort_values("ITEMID")
uom1 = uom1.groupby("ITEMID").count().reset_index()
itemid1 = uom1[uom1.VALUEUOM>1][["ITEMID"]]
itemid0 = uom1[uom1.VALUEUOM<=1][["ITEMID"]]
df_output2 = df_output.merge(itemid0,how="inner",left_on=["ITEMID"],right_on=["ITEMID"])
# itemid with 90%
if len(itemid1) >0:
    uom2 = df_output.merge(itemid1,how="inner",left_on=["ITEMID"],right_on=["ITEMID"])
    uom2 = uom2.groupby(["ITEMID","VALUEUOM"]).count().reset_index()
    uom2 = uom2.merge(uom2.groupby(["ITEMID"]).sum().reset_index(),how="inner",left_on="ITEMID",right_on="ITEMID")
    uom2["Percent"] = uom2.VALUE_x / uom2.VALUE_y

    uom2 = uom2[uom2.Percent>=0.9]
    df_output3 = df_output.merge(uom2[["ITEMID","VALUEUOM"]],how="inner",left_on=["ITEMID","VALUEUOM"],right_on=["ITEMID","VALUEUOM"])
    df_output4 = pd.concat([df_output2,df_output3],axis=0).sort_values(["SUBJECT_ID","HADM_ID","CHARTTIME"])
else:
    df_output4 = df_output2
df_output4.to_csv(os.path.join(mimic3_path2,"OUTPUTEVENTS2.csv"))
print(df_output4.shape)
print("- rows: ",(df_output.shape[0]-df_output4.shape[0]))
```
