# INPUTEVENTS_CV

```python
df_incv = pd.read_csv(os.path.join(mimic3_path,"INPUTEVENTS_CV.csv"),low_memory=False)
df_incv
```

```python
itemid1 = np.loadtxt("./resources/others/itemids_uom_INPUTEVENTS_CV.txt")
print(len(itemid1))
uom2 = df_incv[["ITEMID","AMOUNT","AMOUNTUOM"]].dropna(axis=0,subset = ["AMOUNT", "AMOUNTUOM"]).sort_values("ITEMID")
uom2 = uom2.query(f'{list(itemid1)} in ITEMID')
uom2 = uom2.groupby(["ITEMID","AMOUNTUOM"]).count().reset_index()
uom4 = uom2.merge(uom2.groupby(["ITEMID"]).sum().reset_index(),how="inner",left_on="ITEMID",right_on="ITEMID")
uom4["Percent"] = uom4.AMOUNT_x / uom4.AMOUNT_y
#uom4 = uom4[uom4.Percent>=0.9]
itemid11 = uom4.ITEMID.unique()
print(len(itemid11))
#np.savetxt("./resources/others/itemids_uom_INPUTEVENTS_CV.txt", itemid11, fmt="%d")
uom4
```

```python
incv_map = pd.read_csv( "./resources/itemid_uom_map/inputevents_cv_map.csv")
uom4[uom4.Percent>0.5].merge(incv_map,how="inner",left_on=["ITEMID","AMOUNTUOM"],right_on=["ITEMID","AMOUNTUOM"])
```
